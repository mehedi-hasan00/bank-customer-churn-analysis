```python
# label categorical columns
#for churn
df['churn_label'] = df['churn'].map({0:'Retained',
                      1 : 'Churned'})
# for active member
df['credit_card_label'] = df['credit_card'].map({0 : 'No',
                      1 : 'Yes'})

# for active member
df['active_member_label'] = df['active_member'].map({ 0 : 'Inactive',
                        1 : 'Active'})
```

```python
df.head()
```

```python
df['age'].unique()
```

```python
df.rename(columns = {'estimated_salary' : 'salary',
                    'products_number' : 'num_products'}, inplace = True)
```

```python
df.head()
```

```python
df.drop(columns = 'customer_id', inplace = True)
```

```python
df.head()
```

```python
# create age and balance group
bins = [18,30,40,50,60,100]
labels = ['18-30','31-40','41-50','51-60','60+']
df['age_group'] = pd.cut(
    df['age'],
    bins = bins,
    labels = labels,
    right = True,
    include_lowest= True
)
```

```python
df.head()
```

```python
df['balance_group'] = pd.cut(df['balance'],
                       bins = [-1,0,50000, 100000, 200000, 300000],
                       labels = ['Zero','Low','Medium','High','Very High'])
```

```python
df.head()
```

```python
df['credit_score_group'] = pd.cut(df['credit_score'], 
    bins=[300, 500, 650, 750, 850],
    labels=['Poor', 'Fair', 'Good', 'Excellent'])
```

```python
df.to_csv('bank_churn_cleaned.csv', index=False)
```
